2. Label your x axis "Elements"
3. Label your y axis "Operations"
4. Label your curves or make a legend for the curves
5. Use LaTex notation where possible

```python
import matplotlib.pyplot as plt
import math

elements = range(1, 50)
```

### $y=0x+1$ and label the curve "O(1)"

```python
y1 = [(0 * x + 1) for x in elements]
plt.plot(elements, y1, label = "$O(1)$", color = "red")
plt.legend()
```

### $y=log(x)$ and label the curve "O(log n)"

```python
y2 = [math.log(x) for x in elements]

plt.plot(elements, y2, color = "orange", label = "$O(\log(x))$")
plt.legend()
```

### $y=x$ and label the curve "O(n)"

```python
y3 = elements

plt.plot(elements, y3, color = "yellow", label = "$O(n)$")
plt.legend()
```

### $y=xâˆ—log(x)$ and label it "O(n log n)"

```python
y4 = [(x * (math.log(x))) for x in elements]

plt.plot(elements, y4, color = "green", label = "$O(n\log(n))$")
plt.legend()
```

### $y=x2$ and label it "O(n^2)"

```python
y5 = [x ** 2 for x in elements]

plt.plot(elements, y5, color = "blue", label = "$O(n^2)$")
plt.legend()
```

### $y=2^x$ and label it "O(2^n)"

```python
y6 = [2 ** x for x in elements]

plt.plot(elements, y6, color = "indigo", label = "$O(2^n)$")
plt.legend()
```

### $y=n!$ and label it "O(n!)"

```python
y7 = [math.factorial(x) for x in elements]

plt.plot(elements, y7, color = "violet", label = "$O(n!)$")
plt.legend()
```

### $y=n^n$  and label it "O(n^n)"

```python
y8 = [x ** x for x in elements]

plt.plot(elements, y8, color = "black", label = "$O(n^n)$")
plt.legend()
```

```python
plt.plot(elements, y1, label = "$O(1)$", color = "red")
plt.plot(elements, y2, color = "orange", label = "$O(\log(x))$")
plt.plot(elements, y3, color = "yellow", label = "$O(n)$")
plt.plot(elements, y4, color = "green", label = "$O(n\log(n))$")
plt.plot(elements, y5, color = "blue", label = "$O(n^2)$")
plt.plot(elements, y6, color = "indigo", label = "$O(2^n)$")
plt.plot(elements, y7, color = "violet", label = "$O(n!)$")
plt.plot(elements, y8, color = "black", label = "$O(n^n)$")

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.xlim([0, 20])
plt.ylabel("Operations")
plt.ylim([0, 50])
plt.legend(loc = "best", fontsize = 8)
plt.show()
```
