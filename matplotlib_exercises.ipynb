### Use matplotlib to plot the following equation:  $y=x^2âˆ’x+2$
    1. You'll need to write the code that generates the x and y points.
    2. Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt

x_values = range(0, 14)
y = [(x ** 2 - x + 2) for x in x_values]

plt.plot(x_values, y)
plt.title("$y = x^2 - x + 2$")
plt.annotate("$(0,0)$", xy = (0,0), xytext = (0, 70), arrowprops = {})
```

### Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
#### 1. $y=âˆšx$
#### 2. $y=x^3$
#### 3. $y=2x$
#### 4. $y=1/(x+1)$
    * You can use functions from the math module to help implement some of the equations above.

```python
# 1.
x1_values = range(0, 50)
y1 = [x ** .5 for x in x_values]

plt.plot(x1_values, y1)
plt.title("$sqrt(x)$")
```

```python
# 2.
x2_values = range(0, 50)
y2 = [x ** 3 for x in x_values]

plt.plot(x2_values, y2)
plt.title("$x^3$")
```

```python
# 3.
x3_values = range(0, 50)
y3 = [x * 2 for x in x_values]

plt.plot(x3_values, y3)
plt.title("$2x$")
```

```python
#4. 
x4_values = range(0, 50)
y4 = [(1 / (x + 1)) for x in x_values]

plt.plot(x4_values, y4)
plt.title("$1/(x+1)$")
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(221)
plt.plot(x1_values, y1)
plt.title("$sqrt(x)$")

plt.subplot(222)
plt.plot(x2_values, y2)
plt.title("$x^3$")

plt.subplot(223)
plt.plot(x3_values, y3)
plt.title("$2x$")

plt.subplot(224)
plt.plot(x4_values, y4)
plt.title("$1/(x+1)$")

plt.suptitle("Combined figure")
plt.tight_layout()
plt.show()
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.plot(x1_values, y1, color = "red", label = "$sqrt(x)$")
plt.plot(x2_values, y2, color = "orange", label = "$x^3$")
plt.plot(x3_values, y3, color = "yellow", label = "$2x$")
plt.plot(x4_values, y4, color = "green", label = "$1/(x+1)$")

plt.title("Multiple Equations")
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.legend()
```
